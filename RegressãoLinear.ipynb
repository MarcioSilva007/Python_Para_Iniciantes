```python
#importar pacote para fazer a analise
import numpy as np

#importar pacote para fazer o grafico
import matplotlib.pyplot as plt

x = [1,2,3,4]#variavel idependente
y = [5,8,11,14]#variavel dependente

#polyfit função para fazer a analise de regressão linear o 1 no final
#significa que é de grau1 (polinomio de 1 grau)

fit = np.polyfit(x,y,1)


#funçao polyfit responsável por criar a analise regressão
#guardando o resultado na variavel fit_fn
fit_fn = np.poly1d(fit)

#impressão das constantes
print(fit_fn)

print(round(fit_fn(10),4))

plt.plot(x,y, 'bo', x, fit_fn(x), 'red')
plt.xlim(0,5)
plt.ylim(0,15)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = [0,5,10,20,30,40,50]
y = [1,59.75,356,2411,7666,17621,33776]


#grau 1
fit = np.polyfit(x,y,1,full=True)
fit_fn1 = np.poly1d(fit[0])

print(fit_fn1)
residuo1 = fit[1][0]


#grau 2
fit = np.polyfit(x,y,2,full=True)
fit_fn2 = np.poly1d(fit[0])

print(fit_fn2)
residuo2 = fit[1][0]


#grau 3
fit = np.polyfit(x,y,3,full=True)
fit_fn3 = np.poly1d(fit[0])

print(fit_fn3)
residuo3 = fit[1][0]

print("Polinomio Grau 1. Resíduo:",round(residuo1,2), "Predição para valor igual a 25: ",round(fit_fn1(25),4))
print("Polinomio Grau 2. Resíduo:",round(residuo2,2), "Predição para valor igual a 25: ",round(fit_fn2(25),4))
print("Polinomio Grau 3. Resíduo:",round(residuo3,2), "Predição para valor igual a 25: ",round(fit_fn3(25),4))
plt.plot(x,y, 'o', x, fit_fn1(x), 'brown')
plt.plot(x,y, 'o', x, fit_fn2(x), 'red')
plt.plot(x,y, 'o', x, fit_fn3(x), 'blue')
```

#Regressão Linear Múltipla

```python
import numpy as np
from sklearn import linear_model

#criando matrizes independentes
x_train = np.array([ [1,1,1], [2,3,5], [3,4,4], [4,6,7],[7,8,9],[5,2,1],
 [3,4,5],[3,3,3],[4,4,4],[5,5,5]]);

#matrizes dependentes
y_train = np.array([ 0,3,3,5,12,10,4,4,6,8]);

#criando a varivel model para guardar da função linear_model.LinearRegression() responsável por fazer a regressão
#.fit foi usado para preencher as matrizes x_train e y_train
model = linear_model.LinearRegression().fit(x_train, y_train)

#mostrando o coeficientes de y,x,z,w
#usando str para transformar em string e poder concatenar com o sinal de +
print("Y = " + str(round(model.coef_[0],2))+"X + " + str(round(model.coef_[1],2))
 + "Z + " + str(round(model.coef_[2],2)) + "W + "+ str(round(model.intercept_,2)) )

#mostrando a erro médio da analise, função men  responsável por calcular a media
#funão predict responsável por calcular a predição
print("Erro médio quadrado: %.2f" % np.mean((model.predict(x_train) - y_train)
** 2))


print(model.predict([[10,10,10]]))
```
